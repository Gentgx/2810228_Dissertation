```python
# å®šä¹‰è®­ç»ƒå‡½æ•°
def train_yolo(
    # é»˜è®¤æ¨¡å‹è®­ç»ƒå‚æ•°
    epochs=100, # è®­ç»ƒè½®æ•°ï¼Œå³epochæ•°
    batch=16, # æ‰¹å¤§å°ï¼Œå³æ¯æ¬¡è®­ç»ƒè¾“å…¥çš„å›¾åƒæ•°
    imgsz=640, # è¾“å…¥å›¾åƒå°ºå¯¸ï¼ŒYOLOä¼šè‡ªåŠ¨ç¼©æ”¾åˆ°è¯¥å¤§å°
    optimizer='SGD', # ä¼˜åŒ–å™¨ï¼Œæ”¯æŒSGDã€Adamã€AdamWç­‰
    lr0=0.01, # åˆå§‹å­¦ä¹ ç‡
    patience=50, # æ—©åœç­–ç•¥çš„è€å¿ƒå€¼ï¼Œå³å¤šå°‘epochå†…æ²¡æœ‰æå‡å°±åœæ­¢è®­ç»ƒ
    weight_decay=0.0005, # æƒé‡è¡°å‡ï¼Œæ˜¯æ­£åˆ™åŒ–å‚æ•°ï¼Œç”¨äºé˜²æ­¢è¿‡æ‹Ÿåˆ
    augment=False, # æ˜¯å¦å¯ç”¨æ•°æ®å¢å¼ºï¼Œè¿™é‡Œé»˜è®¤å…³é—­

    # æ•°æ®é›†é…ç½®yamlæ–‡ä»¶ï¼ŒåŒ…å«è®­ç»ƒ/éªŒè¯/æµ‹è¯•é›†è·¯å¾„åŠç±»åˆ«ä¿¡æ¯
    data_path="/content/drive/MyDrive/tobacco_data/yolo_yaml.yaml",
    # é¢„è®­ç»ƒæ¨¡å‹æƒé‡
    model="yolov8n.pt",

    # è®­ç»ƒç»“æœä¿å­˜è·¯å¾„çš„ä¸Šçº§ç›®å½•
    project="/content/drive/MyDrive/tobacco_data/runs/hyperparameter_tuning",
    # æœ¬æ¬¡å®éªŒçš„å­æ–‡ä»¶å¤¹åç§°
    name="yolov8n_pt"
):
    # åŠ è½½YOLOæ¨¡å‹ï¼ŒåŸºäºæŒ‡å®šçš„é¢„è®­ç»ƒæƒé‡
    model = YOLO(model)
    # å¼€å§‹è®­ç»ƒ
    model.train(
        data=data_path,
        epochs=epochs,
        batch=batch,
        imgsz=imgsz,
        optimizer=optimizer,
        lr0=lr0,
        patience=patience,
        augment=augment,
        weight_decay=weight_decay,
        project=project,
        name=name
    )
```

```python
# è°ƒç”¨YOLOå‡½æ•°ï¼Œå¼€å§‹è®­ç»ƒyolov8n.pt
train_yolo(
    epochs=150, # è‡ªå®šä¹‰è®­ç»ƒè½®æ•°
    model="yolov8n.pt",
    project="/content/drive/MyDrive/tobacco_data/runs/detect",
    name="yolov8n_pt"
)
```

```python
# åŠ è½½å·²ç»è®­ç»ƒå¥½çš„YOLOv8æ¨¡å‹ï¼Œè¿™é‡Œä½¿ç”¨çš„æ˜¯ä¹‹å‰è®­ç»ƒä¿å­˜çš„æœ€ä½³æƒé‡æ–‡ä»¶
model = YOLO("/content/drive/MyDrive/tobacco_data/runs/detect/yolov8n_pt/weights/best.pt")

# åœ¨æµ‹è¯•é›†ä¸Šå¯¹æ¨¡å‹è¿›è¡ŒéªŒè¯ï¼Œè®¡ç®—æ£€æµ‹æŒ‡æ ‡
metrics = model.val(
    # æŒ‡å®šæ•°æ®é›†é…ç½®æ–‡ä»¶YAML
    data="/content/drive/MyDrive/tobacco_data/yolo_yaml.yaml",
    # ä½¿ç”¨æµ‹è¯•é›†
    split="test"
)
```

```python
# å®šä¹‰é¢„æµ‹å‡½æ•°ï¼Œç”¨äºåŠ è½½è®­ç»ƒå¥½çš„YOLOæ¨¡å‹å¹¶åœ¨ç»™å®šæ•°æ®ä¸Šè¿›è¡Œé¢„æµ‹
def run_prediction(model_path, source_path, save=True, show=True):
    # åŠ è½½æŒ‡å®šè·¯å¾„ä¸‹çš„YOLOæ¨¡å‹æƒé‡æ–‡ä»¶
    model = YOLO(model_path)
    model.predict(
        source=source_path, # å¾…é¢„æµ‹çš„å›¾åƒæˆ–æ–‡ä»¶å¤¹è·¯å¾„
        classes=1, # æŒ‡å®šæ£€æµ‹çš„ç±»åˆ«ï¼Œ1è¡¨ç¤ºç­¾åç±»åˆ«ï¼Œé¿å…æ£€æµ‹æ— å…³å¯¹è±¡
        hide_labels=True, # åœ¨é¢„æµ‹å›¾åƒä¸­éšè—ç±»åˆ«æ ‡ç­¾
        hide_conf=True, # åœ¨é¢„æµ‹å›¾åƒä¸­éšè—ç½®ä¿¡åº¦åˆ†æ•°
        line_thickness=2,
        project='/content/drive/MyDrive/tobacco_data/runs/predict', # æŒ‡å®šé¢„æµ‹ç»“æœä¿å­˜çš„é¡¹ç›®ç›®å½•
        name='yolov8n_pt', # è®¾ç½®ç»“æœä¿å­˜çš„å­ç›®å½•åç§°
        save=save, # æ§åˆ¶æ˜¯å¦ä¿å­˜ç»“æœ
        show=show # æ§åˆ¶æ˜¯å¦åœ¨å±å¹•å®æ—¶å±•ç¤ºé¢„æµ‹è¿‡ç¨‹
    )
```

```python
# è°ƒç”¨é¢„æµ‹å‡½æ•°
run_prediction(
    # æŒ‡å®šä½¿ç”¨YOLOv8næ¨¡å‹çš„æœ€ä½³æƒé‡best.pt
    model_path='/content/drive/MyDrive/tobacco_data/runs/detect/yolov8n_pt/weights/best.pt',
    # æŒ‡å®šè¾“å…¥æºï¼Œä¹Ÿå°±æ˜¯è¦é¢„æµ‹çš„æ–‡æ¡£å›¾åƒ
    #source_path='/content/drive/MyDrive/tobacco_data/yolo_format/images/test/aah97e00-page02_1.jpg'
    source_path='/content/drive/MyDrive/tobacco_data/yolo_format/images/test'
)
```
